# Tarea 5: Golpe de ariete

Datos del problema:

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
q10=0.3
q20=0.3
q30=0.6
L1=1000.
L2=800.
L3=500.
d1=0.25
d2=0.35
d3=0.35
f=0.016
a1=860.
a2=960.
a3=940.
g=9.81
dt=0.01
dx1=dt*a1
dx2=dt*a2
dx3=dt*a3
z=390.
T=20.
nt=int(T/dt)   
```

Arreglos para guardar los datos:

```python
x1 = np.arange(0.,L1+dx1,dx1)
x2 = np.arange(0.,L2+dx2,dx2)
x3 = np.arange(0.,L3+dx3,dx3)
h1 = np.zeros((nt+1,x1.shape[0]))
h2 = np.zeros((nt+1,x2.shape[0]))
h3 = np.zeros((nt+1,x3.shape[0]))
v1 = np.zeros((nt+1,x1.shape[0]))
v2 = np.zeros((nt+1,x2.shape[0]))
v3 = np.zeros((nt+1,x3.shape[0]))
```

Cálculo de variables en régimen permanente

```python
v10=q10/(np.pi*d1**2/4)
print 'v10=',v10
v20=q20/(np.pi*d2**2/4)
print 'v20=',v20
v30=q30/(np.pi*d3**2/4)
print 'v30=',v30
Hb1=z+f*L1*v10**2/(2*g*d1)+f*L3*v30**2/(2*g*d3)
print 'Hb1=',Hb1
Hb2=z+f*L2*v20**2/(2*g*d2)+f*L3*v30**2/(2*g*d3)
print 'Hb2=',Hb2
```

Condiciones iniciales

```python
h1[0,:] = Hb1-f*x1*v10**2/(2*g*d1)
v1[0,:] = v10
h2[0,:] = Hb2-f*x2*v20**2/(2*g*d2)
v2[0,:] = v20
h3[0,:] = z+f*(L3-x3)*v30**2/(2*g*d3)
v3[0,:] = v30
```

Función de cierre de la válvula. Se evalúan tres casos para las líneas 1 y 2

```python
tc1=0.5*min(L1/a1,L2/a2)
tc2=2.*min(L1/a1,L2/a2)
tc3=10.*min(L1/a1,L2/a2)
def v_valv(t,tc,v0):
    return max(v0*(1.-t/tc),0.)
```

Algoritmo de resolución

Caso 1: tc=tc1

(*)Nota: Los casos restantes se obtienen al reemplazar tc1 por tc2 y tc3 respectivamente. Al cambiar el tc se ejecutan las líneas siguientes nuevamente, y se obtienen los respectivos gráficos que son reportados en el informe.

```python
for n in range(nt):
    #Inicio de las líneas 1 y 2
    i = 0
    v1[n+1,i] = v_valv((n+1)*dt,tc1,v10)
    h1[n+1,i] = h1[n,i+1] + a1/g*(v1[n+1,i]-v1[n,i+1] + dt*f*v1[n,i+1]*np.abs(v1[n,i+1])/(2.*d1))
    v2[n+1,i] = v_valv((n+1)*dt,tc1,v20)
    h2[n+1,i] = h2[n,i+1] + a2/g*(v2[n+1,i]-v2[n,i+1] + dt*f*v2[n,i+1]*np.abs(v2[n,i+1])/(2.*d2))

    #Término de la línea 3
    i = -1
    h3[n+1,i] = z
    v3[n+1,i] = v3[n,i-1]+g/a3*(h3[n,i-1]-h3[n+1,i])-dt*f*v3[n,i-1]*np.abs(v3[n,i-1])/(2.*d3)
    
    #Nodo que une las líneas 1, 2 y 3
    C1=v1[n,-2]+g*h1[n,-2]/a1-dt*f*v1[n,-2]*np.abs(v1[n,-2])/(2*d1)
    C2=g/a1
    C3=v2[n,-2]+g*h2[n,-2]/a2-dt*f*v2[n,-2]*np.abs(v2[n,-2])/(2*d2)
    C4=g/a2
    C5=v3[n,1]-g*h3[n,1]/a3-dt*f*v3[n,1]*np.abs(v3[n,1])/(2*d3)
    C6=g/a3
    H=C1*(np.pi*d1**2/4)-C3*(np.pi*d2**2/4)-C5*(np.pi*d3**2/4)/(C2*(np.pi*d1**2/4)+C4*(np.pi*d2**2/4)+C6*(np.pi*d3**2/4))
    h1[n+1,-1]=H
    h2[n+1,-1]=H
    h3[n+1,0]=H
    v1[n+1,-1]=C1-C2*h1[n+1,-1]
    v2[n+1,-1]=C3-C4*h2[n+1,-1]
    v3[n+1,0]=C5+C6*h3[n+1,0]
    
    #interior de la tubería
    h1[n+1,1:-1] = g/a1*(h1[n,2:]+h1[n,:-2]) 
    h1[n+1,1:-1]+= - (v1[n,2:]-v1[n,:-2]) 
    h1[n+1,1:-1]+= - dt*f/(2.*d1) * (v1[n,:-2]*np.abs(v1[n,:-2]) - v1[n,2:]*np.abs(v1[n,2:]))
    h1[n+1,1:-1]*= 0.5*a1/g
    
    h2[n+1,1:-1] = g/a2*(h2[n,2:]+h2[n,:-2]) 
    h2[n+1,1:-1]+= - (v2[n,2:]-v2[n,:-2]) 
    h2[n+1,1:-1]+= - dt*f/(2.*d2) * (v2[n,:-2]*np.abs(v2[n,:-2]) - v2[n,2:]*np.abs(v2[n,2:]))
    h2[n+1,1:-1]*= 0.5*a2/g
    
    h3[n+1,1:-1] = g/a3*(h3[n,2:]+h3[n,:-2]) 
    h3[n+1,1:-1]+= - (v3[n,2:]-v3[n,:-2]) 
    h3[n+1,1:-1]+= - dt*f/(2.*d3) * (v3[n,:-2]*np.abs(v3[n,:-2]) - v3[n,2:]*np.abs(v3[n,2:]))
    h3[n+1,1:-1]*= 0.5*a3/g
    
    v1[n+1,1:-1] = v1[n,2:] + v1[n,:-2] - g/a1*(h1[n,2:]-h1[n,:-2]) - dt*f/(2*d1)*(v1[n,:-2]*np.abs(v1[n,:-2]) + v1[n,2:]*np.abs(v1[n,2:]))
    v1[n+1,1:-1] *=0.5
    
    v2[n+1,1:-1] = v2[n,2:] + v2[n,:-2] - g/a2*(h2[n,2:]-h2[n,:-2]) - dt*f/(2*d2)*(v2[n,:-2]*np.abs(v2[n,:-2]) + v2[n,2:]*np.abs(v2[n,2:]))
    v2[n+1,1:-1] *=0.5
    
    v3[n+1,1:-1] = v3[n,2:] + v3[n,:-2] - g/a3*(h3[n,2:]-h3[n,:-2]) - dt*f/(2*d3)*(v3[n,:-2]*np.abs(v3[n,:-2]) + v3[n,2:]*np.abs(v3[n,2:]))
    v3[n+1,1:-1] *=0.5
```
